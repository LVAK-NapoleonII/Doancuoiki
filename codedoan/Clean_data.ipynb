```python
import pandas as pd
```

```python
df = pd.read_csv("Fixed_news_dataset.csv")
df
```

```python
df = df.drop_duplicates() #xóa dữ liệu trùng
df
```

```python
df = df.drop(columns = "crawled_at")
df
```

```python
df["content"] = df["content"].str.lstrip("*")
df["content"] = df["content"].str.replace("\"", "")
df
```

```python
df=df.fillna('')#loại bỏ NaN
df
```

```python
for x in df.index:
    if df.loc[x, "author"] == '':
        df.drop(x, inplace=True)
for x in df.index:
    if df.loc[x, "content"] == '':
        df.drop(x, inplace=True)
for x in df.index:
    if df.loc[x, "picture_count"] == '':
        df.drop(x, inplace=True)
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.to_csv("cleaned_data.csv")
```
